### Outliers

 Mengecek outliers bertujuan untuk membedakan antara "data yang salah (noise)" dan "data yang unik/berharga", lalu menanganinya dengan tepat (dihapus, diubah, atau dianalisis terpisah).

 Namun di outliers dataset ini tidak ada yang perlu ditangani.

 Lumayan bersih... yah datanya.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## Outliers

"Outliers adalah nilai yang jauh berbeda dari nilai lainnya dalam dataset. Nilai Outlier bisa jauh lebih rendah atau lebih tinggi. Outlier bisa terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lain yang tidak terduga." - Mengutik dari penjelasan Pertemuan 3 AVD

Disini terlihat tidak ada outlier sedikit pun, jadi tidak ada yang perlu ditangani. Dikarenakan tidak ada nilai yang jauh dari nilai-nilai lainnya di dataset.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
def categorize_intensity(hours):
    if hours > 10:
        return 'Berat'    # Penggunaan sangat tinggi
    elif hours > 6:
        return 'Sedang'   # Penggunaan wajar
    else:
        return 'Ringan'   # Penggunaan minimal

df['Intencity'] = df['total_screen_time'].apply(categorize_intensity)

df
```

*   **Insight:** Heatmap dan Scatter Plot menunjukkan korelasi positif yang sangat kuat (0.8+) antara waktu media sosial dan total screen time. Media sosial bukan sekadar aktivitas sampingan, melainkan pendorong utama durasi online.
*   **Tindakan:** Alokasikan minimal 60% anggaran pemasaran pada iklan media sosial. Karena penggunaan media sosial berbanding lurus dengan waktu layar, retensi di media sosial akan secara otomatis meningkatkan eksposur merek secara keseluruhan.

### **3. Preferensi Perangkat & Koneksi**
*   **Insight:** **Tablet** secara mengejutkan menjadi perangkat yang sangat populer di beberapa segmen, bersaing ketat dengan Mobile. Selain itu, **WiFi** tetap menjadi jenis koneksi dominan.
*   **Tindakan:** Jangan hanya melakukan optimasi 'Mobile-First', tapi juga 'Tablet-Friendly'. Pastikan aplikasi atau situs web Anda responsif pada resolusi tablet untuk menangkap peluang dari segmen pengguna ini.

### **4. Pola Distribusi & Kesehatan Digital**
*   **Insight:** Histogram menunjukkan distribusi screen time yang luas, dengan puncaknya di angka 6-10 jam. Boxplot mengungkap adanya variabilitas penggunaan yang cukup lebar di kelompok usia muda.
*   **Tindakan:** Terapkan fitur *Digital Wellbeing*. Memberikan laporan penggunaan atau saran istirahat kepada pengguna dengan screen time >10 jam akan membangun loyalitas jangka panjang dan citra perusahaan yang bertanggung jawab secara sosial.

### **5. Tren Waktu (Temporal)**
*   **Insight:** Data menunjukkan fluktuasi harian dalam pengumpulan data, mengindikasikan adanya periode 'peak' tertentu dalam penggunaan internet harian.
*   **Tindakan:** Lakukan *flash sale* atau perilisan konten utama pada periode puncak tersebut untuk memaksimalkan *engagement rate* dan konversi penjualan.
